```python
def get_info_on_dataset(_df, _full = True):

    CONST_IN_PERCENTAGES = float(100.0)
    
    try:
        _df.info()
        df_review_info = pd.DataFrame({'Not_Valid_Values_in %': (_df.isna().sum()/len(_df))*CONST_IN_PERCENTAGES} )
        print("")
        print("Number of duplicated rows in the dataset: {}".format(_df.duplicated().sum()))
        print("")
        display(df_review_info.sort_values(by='Not_Valid_Values_in %', ascending=False))
    except:
        pprint("Can't provide review on given dataset")
    
    pass
```

```python
def plot_hist(_x_data, 
              _bins = 100, 
              _x_ticks_range = [],
              _y_ticks_range = [],
              _outliers_removed = False,
              _kde = True,
              _x_axis_label = "x", 
              _y_axis_label = "count", 
              _width = 12, 
              _height = 8, 
              _fontsize=12):
    
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
        
    try:
        
        if(_outliers_removed):
            
            Q1 = _x_data.quantile(Q1_QUANTILE_VALUE_CONST)
            Q3 = _x_data.quantile(Q3_QUANTILE_VALUE_CONST)
            IQR = Q3 - Q1 
            
            outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
            outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR
            
            _x_data = _x_data[ (_x_data >= outliers_low_bound) & (_x_data <= outliers_upper_bound) ]
    
        fig = plt.figure(figsize=(_width, _height))
        sns.histplot(x=_x_data, bins=_bins, kde=_kde)
        plt.xlabel(_x_axis_label, fontsize=_fontsize)
        plt.ylabel(_y_axis_label, fontsize=_fontsize)
                
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)
        
    except:
        pprint("Can't provide histogram plotting")
        
    pass
```

```python
def plot_df_coll_matrix(_df, 
                        _columns, 
                        _fiture__height = 15, 
                        _fiture_width = 15, 
                        _cell_annotation_font_size = 15):
  
    try:
        corr = _df[_columns].corr()
        fig, ax = plt.subplots(figsize=(_fiture__height, _fiture_width))
        sns.heatmap(corr, annot=True, linewidths=.5, ax=ax, annot_kws={'size': _cell_annotation_font_size})
    except:
        pprint("Can't provide pairwise correlation plotting")
    
    pass
```

```python
def plot_feature_target_hists(_df,
                             _feature_column,
                             _ml_target_column,
                             _bins = 50, 
                             _figure_width = 20, 
                             _figure_height = 12, 
                             _fontsize=15):

    try:
        
        plt.figure(figsize=(_figure_width, _figure_height))

        sns.histplot(data=_df.query("{} == 0".format(_ml_target_column)), 
                     x=_feature_column, 
                     color="crimson", 
                     label= str(_feature_column  + ": {} - {}".format(_ml_target_column, 0)), 
                     bins = _bins, 
                     kde=True)
        
        sns.histplot(data=_df.query("{} == 1".format(_ml_target_column)), 
                     x=_feature_column, 
                     color="c", 
                     label=str(_feature_column  + ": {} - {}".format(_ml_target_column, 1)), 
                     bins = _bins, 
                     kde=True)

        plt.xlabel("Feature - {}".format(_feature_column), fontsize=_fontsize)
        plt.ylabel('', fontsize=_fontsize)
        plt.yticks(fontsize=_fontsize)
        plt.xticks(fontsize=_fontsize)
        

        mean_f0_value = _df.loc[_df[_ml_target_column] == 0, _feature_column].mean()  
        median_f0_value = _df.loc[_df[_ml_target_column] == 0, _feature_column].median() 

        mean_f1_value = _df.loc[_df[_ml_target_column] == 1, _feature_column].mean()  
        median_f1_value = _df.loc[_df[_ml_target_column] == 1, _feature_column].median() 
        
        plt.axvline(x=mean_f0_value, color='black', lw = 5, ls = ':', 
                    label = "mean(ml_target - 0): {}".format(round(mean_f0_value, 2)))
        plt.axvline(x=median_f0_value, color='black', lw = 5, ls = ':', 
                    label = "median(ml_target - 0): {}".format(round(median_f0_value, 2)))

        plt.axvline(x=mean_f1_value, color='blue', lw = 5, ls = ':', 
                    label = "mean(ml_target - 1): {}".format(round(mean_f1_value, 2)))
        plt.axvline(x=median_f1_value, color='blue', lw = 5, ls = ':', 
                    label = "median(ml_target - 1): {}".format(round(median_f1_value, 2)))
        
        plt.legend(fontsize = '15', title_fontsize = "10")
        plt.show() 

        
    except:
        pprint("Can't provide boxplot plotting")
        
    pass
```

```python
def plot_catplot(_df_data, 
                 _x_col_title,
                 _y_col_title,
                 _x_ticks_range = [],
                 _y_ticks_range = [],
                 _x_axis_label = "", 
                 _y_axis_label = "", 
                 _height= 1, 
                 _aspect = 1, 
                 _order = [],
                 _fontsize=12):

    try:
                    
        if _order:
            sns.catplot(x=_x_col_title, y=_y_col_title, kind="box", 
                        data=_df_data, height=_height, aspect=_aspect, palette='dark', order=_order)
        else:
            sns.catplot(x=_x_col_title, y=_y_col_title, kind="box", 
                        data=_df_data, height=_height, aspect=_aspect, palette='dark')
            
        plt.xlabel(_x_axis_label, fontsize=_fontsize)
        plt.ylabel(_y_axis_label, fontsize=_fontsize)
        
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)

    except:
        pprint("Can't provide catplot plotting")
        
    pass
```

В дальшейшем для работы нам нужно преобразовать категориальные признаки в численные. Для этого применим метод прямого кодирования One-Hot Encoding к двум столбцам `geography` и `gender`.

### LR - Logistic Regression Classifier  - Логистическая регрессия

Оптимизируемые параметры алгоритма и используемые при этом значения: 
- **C:**          относительная сила регуляризации
- **solver:**     решатель для подбора модели
- **tol:**        допуск для остановки процедуры

```python
c = [0.01, 0.1, 0.5, 1.0, 2.0, 4.0, 7.0, 10.0, 100.0] 
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'] 
tol = [1e-6, 1e-4, 1e-2] 
```

```python
df_lr_results = pd.DataFrame(columns=['c',
                                      'solver',
                                      'tol',
                                      'accuracy',
                                      'auc_roc',
                                      'f1_score'])                            
    

start = time.time()
print("Optimal LR parameter values computation started...")

for _c in c:
    for _solver in solver:
        for _tol in tol:
            
                model_lr = LogisticRegression(C=_c,
                                              solver=_solver,
                                              tol=_tol,
                                              random_state=42)
                
                model_lr.fit(df_features_training, df_target_training.values.ravel())
                        
                _actual_pred = model_lr.predict(df_features_validation)
                _actual_accuracy_value = accuracy_score(df_target_validation, _actual_pred)
                _actual_f1_value = f1_score(df_target_validation, _actual_pred)
                        
                _actual_predict_proba = model_lr.predict_proba(df_features_validation)
                _actual_predict_proba_values = _actual_predict_proba[:, 1]
                _actual_auc_roc_value = roc_auc_score(df_target_validation, _actual_predict_proba_values)
                
                df_actual_res = pd.DataFrame.from_records([ {'c': _c,
                                                             'solver': _solver,
                                                             'tol': _tol,
                                                             'accuracy': round(_actual_accuracy_value, 3),
                                                             'auc_roc': round(_actual_auc_roc_value, 3),
                                                             'f1_score': round(_actual_f1_value, 3)}])
                        
                df_lr_results = pd.concat([df_lr_results, df_actual_res])                     
                        
print("Optimal LR parameter values computation finished...")      

end = time.time()
print("Execution time {} sec.".format( round(end - start,2) ))

df_lr_results.sort_values(by='f1_score', ascending=False).head(5)
```

### RF - Random Forest Classifier - Случайный лес

Оптимизируемые параметры алгоритма и используемые при этом значения: 
- **n_estimators:** число «деревьев» в «случайном лесу»
- **max_features:** число признаков для выбора расщепления
- **max_depth:** максимальная глубина деревьев
- **min_samples_split:** минимальное число объектов, необходимое для того, чтобы узел дерева мог бы расщепиться
- **min_samples_leaf:** минимальное число объектов в листьях
- **bootstrap:** использование для построения деревьев подвыборки с возвращением

```python
bootstrap = [True, False] 
min_samples_leaf = [1, 2, 3, 4] 
n_estimators = [10, 20, 50] 
min_samples_split = [2, 4, 6, 8, 10, 12] 
max_features = ['auto', 'sqrt', 'log2'] 
max_depth = [6, 8, 10, 12, 14, 16, 18, 20] 
```

```python
df_rf_results = pd.DataFrame(columns=['n_estimators',
                                      'max_features',
                                      'max_depth',
                                      'min_samples_split',
                                      'min_samples_leaf',
                                      'bootstrap',
                                      'accuracy',
                                      'auc_roc',
                                      'f1_score'])

start = time.time()
print("Optimal RF parameter values computation started...")

for _bootstrap in bootstrap:
    for _min_samples_leaf in min_samples_leaf:
        for _n_estimators in n_estimators:
            for _min_samples_split in min_samples_split:
                for _max_features in max_features:
                    for _max_depth in max_depth:
                        
                        model_rf = RandomForestClassifier(n_jobs=-1, 
                                                          n_estimators       = _n_estimators, 
                                                          max_features       = _max_features,
                                                          max_depth          = _max_depth, 
                                                          min_samples_split  = _min_samples_split,
                                                          min_samples_leaf   = _min_samples_leaf,
                                                          bootstrap          = _bootstrap, 
                                                          random_state=424)
                       
                        model_rf.fit(df_features_training, df_target_training.values.ravel())
                        
                        _actual_pred = model_rf.predict(df_features_validation)
                        _actual_accuracy_value = accuracy_score(df_target_validation, _actual_pred)
                        _actual_f1_value = f1_score(df_target_validation, _actual_pred)
                        
                        _actual_predict_proba = model_rf.predict_proba(df_features_validation)
                        _actual_predict_proba_values = _actual_predict_proba[:, 1]
                        _actual_auc_roc_value = roc_auc_score(df_target_validation, _actual_predict_proba_values)

                        df_actual_res = pd.DataFrame.from_records([ {'n_estimators': _n_estimators,
                                                                     'max_features': _max_features,
                                                                     'max_depth': _max_depth,
                                                                     'min_samples_split': _min_samples_split,
                                                                     'min_samples_leaf': _min_samples_leaf,
                                                                     'bootstrap': _bootstrap,
                                                                     'accuracy': round(_actual_accuracy_value, 3),
                                                                     'auc_roc': round(_actual_auc_roc_value, 3),
                                                                     'f1_score': round(_actual_f1_value, 3)}])
                        
                        df_rf_results = pd.concat([df_rf_results, df_actual_res])                     
                        
print("Optimal RF parameter values computation finished...")      

end = time.time()
print("Execution time {} sec.".format( round(end - start, 2) ))

df_rf_results.sort_values(by='f1_score', ascending=False).head(5)
```

### GB - Gradient Boosting Classifier  - Градиентный бустинг

Оптимизируемые параметры алгоритма и используемые при этом значения:
- **learning_rates:**     скорость обучения
- **n_estimators:**       число итераций
- **min_samples_split:**  минимальное число объектов, при котором происходит расщепление
- **max_features:**       количество признаков, учитываемых алгоритмом для построения расщепления в дереве
- **max_depth:**          максимальное число верхних точек в дереве

```python
learning_rates = [0.25, 0.1, 0.05, 0.01]
n_estimators = [50, 100, 150, 200]
min_samples_split = [2, 4, 6, 8]  
max_features = ['auto', 'sqrt', 'log2']
max_depth = [2, 4, 6, 8, 10] 
```

```python
df_gb_results = pd.DataFrame(columns=['n_estimators',
                                      'learning_rate',
                                      'max_depth',
                                      'min_samples_split',
                                      'max_features',
                                      'accuracy',
                                      'auc_roc',
                                      'f1_score'])
start = time.time()
print("Optimal GBoosting parameter values computation started...")

for _learning_rate in learning_rates:
    for _max_depth in max_depth:
        for _n_estimators in n_estimators:
            for _min_samples_split in min_samples_split:
                for _max_features in max_features:
                    
                    model_gb = GradientBoostingClassifier(n_estimators       = _n_estimators, 
                                                          learning_rate      = _learning_rate,
                                                          max_depth          = _max_depth, 
                                                          min_samples_split  = _min_samples_split,
                                                          max_features       = _max_features,
                                                          random_state=424)
                       
                    model_gb.fit(df_features_training, df_target_training.values.ravel())
                        
                    _actual_pred = model_gb.predict(df_features_validation)
                    _actual_accuracy_value = accuracy_score(df_target_validation, _actual_pred)
                    
                    _actual_predict_proba = model_gb.predict_proba(df_features_validation)
                    _actual_predict_proba_values = _actual_predict_proba[:, 1]
                    _actual_auc_roc_value = roc_auc_score(df_target_validation, _actual_predict_proba_values)

                    df_actual_res = pd.DataFrame.from_records([{'n_estimators': _n_estimators,
                                                                'max_features': _max_features,
                                                                'max_depth': _max_depth,
                                                                'min_samples_split': _min_samples_split,
                                                                'learning_rate': _learning_rate,
                                                                'accuracy': round(_actual_accuracy_value, 3),
                                                                'auc_roc': round(_actual_auc_roc_value, 3),
                                                                'f1_score': round(_actual_f1_value, 3)}])
                        
                    df_gb_results = pd.concat([df_gb_results, df_actual_res])                     
                        
print("Optimal GBoosting parameter values computation finished...")      

end = time.time()
print("Execution time {} sec.".format( round(end - start, 2) ))

df_gb_results.sort_values(by='f1_score', ascending=False).head(5)
```

### Использование undersampling

Чтобы сбалансировать классы для обучения моделей попробуем уменьшить объекты большего класса. <br>
Разделим обучающую выборку на объекты по классам, отбросим случайным образом часть объектов из большего класса, создадим новую выборку и перемешаем данные.<br>
На полученной новой обучающей выборке обучим наши модели, проверим их качество и сравним результаты.

```python
def downsample(features, target, fraction):
    
    features_zeros = features[(target == 0).values]
    features_ones = features[(target == 1).values]
    target_zeros = target[(target == 0).values]
    target_ones = target[(target == 1).values]
    
    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)

 
    return features_downsampled, target_downsampled
```

```python
df_features_training_undersampled, df_target_training_undersampled = downsample(df_features_training, 
                                                                                df_target_training, 0.255)
```

```python
print("Размерность обучающего датасета, количество различных элементов в классах до процедуры undersampling:")
print(df_features_training.shape)
print(df_target_training.value_counts())
print("")
print("Размерность обучающего датасета, количество различных элементов в классах после процедуры undersampling:")
print(df_features_training_undersampled.shape)
print(df_target_training_undersampled.value_counts())
```

```python
df_gb_results = pd.DataFrame(columns=['n_estimators',
                                      'learning_rate',
                                      'max_depth',
                                      'min_samples_split',
                                      'max_features',
                                      'accuracy',
                                      'auc_roc',
                                      'f1_score'])
start = time.time()
print("Optimal GBoosting parameter values computation started...")

for _learning_rate in learning_rates:
    for _max_depth in max_depth:
        for _n_estimators in n_estimators:
            for _min_samples_split in min_samples_split:
                for _max_features in max_features:
                    
                    model_gb = GradientBoostingClassifier(n_estimators       = _n_estimators, 
                                                          learning_rate      = _learning_rate,
                                                          max_depth          = _max_depth, 
                                                          min_samples_split  = _min_samples_split,
                                                          max_features       = _max_features,
                                                          random_state=424)
                       
                    model_gb.fit(df_features_training_undersampled, df_target_training_undersampled.values.ravel())
                        
                    _actual_pred = model_gb.predict(df_features_validation)
                    _actual_accuracy_value = accuracy_score(df_target_validation, _actual_pred)
                    
                    _actual_predict_proba = model_gb.predict_proba(df_features_validation)
                    _actual_predict_proba_values = _actual_predict_proba[:, 1]
                    _actual_auc_roc_value = roc_auc_score(df_target_validation, _actual_predict_proba_values)

                    df_actual_res = pd.DataFrame.from_records([{'n_estimators': _n_estimators,
                                                                'max_features': _max_features,
                                                                'max_depth': _max_depth,
                                                                'min_samples_split': _min_samples_split,
                                                                'learning_rate': _learning_rate,
                                                                'accuracy': round(_actual_accuracy_value, 3),
                                                                'auc_roc': round(_actual_auc_roc_value, 3),
                                                                'f1_score': round(_actual_f1_value, 3)}])
                        
                    df_gb_results = pd.concat([df_gb_results, df_actual_res])                     
                        
print("Optimal GBoosting parameter values computation finished...")      

end = time.time()
print("Execution time {} sec.".format( round(end - start, 2) ))

df_gb_results.sort_values(by='f1_score', ascending=False).head(5)
```
