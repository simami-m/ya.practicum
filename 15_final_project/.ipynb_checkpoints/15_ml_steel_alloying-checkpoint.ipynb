# Промышленность — предсказание температуры стали.
<br>

**Цель проекта** - для оптимизации производственных расходов, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Необходимо построить модель, которая предскажет температуру стали.


**Описание процесса обработки стали**<br>

На предприятии сталь обрабатывают в агрерате "ковш-печь" (или же иные названия: дуговая ковшовая печь, АКОС - агрерат комплексной обработки стали). <br>

Сам ковш вместимостью около **100 тонн**. 
Для того, чтобы ковш мог выдерживать высокие температуры его изнутри облицовывают огнеупорным кирпичом.<br>

В предварительно разогретый ковш заливают металл из сталеразливочного ковша.<br>
При переливе стали как можно тщательнее отделяется шлак, образовавшийся в сталеплавильной печи и имеющий повышенное содержание фосфора (лучше, если в сталеплавильной печи не прошла стадия раскисления во избежание обратного перехода фосфора из шлака в сталь). 
Далее в ковш вводят ферросплавы, требуемые для доводки стали до нужного химического состава, и вводят шлакообразующие.<br>
После этого ковш помещают под водоохлаждаемый свод, через отверстие в котором проходят графитированные электроды, и опускают свод на упоры ковша. Электроды опускаются ниже верхнего уровня шлака для экранирования футеровки от излучения дуг и начинается дуговой нагрев. <br>

Скорость нагрева стали в ковшовых печах в зависимости от технологии обработки металла обычно составляет **3-5 °С/мин**.<br>
Желательный уровень температуры металла, поддерживаемой в ковше ~ **1600-1670 °С** (при этом температура плавления стали от 1350 до 1535 °С) <br>

Для исключения попадания в рабочее пространство кислорода и поддержания восстановительной атмосферы под свод дополнительно вводят инертный газ. <br>
Одновременно с нагревом или немного раньше начинают продувку металла аргоном через пористую пробку в днище ковша. Благодаря этому образующийся основной шлак интенсивно перемешивается с металлом, ускоряя процессы раскисления стали и удаления серы.<br>

Через технологическое отверстие в своде проводятся замеры температуры, отбор проб, наблюдение за ходом процесса и добавка легирующих для точной доводки химического состава стали. <br>

Полная длительность обработки стали в этой ковшовой печи **60-90 минут**, из них в течение **10-30 минут** циклически может происходить дуговой нагрев.<br>

После окончания каждого цикла нагрева электроды поднимаются. <br>
Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.<br>

После окончания рафинирующих операций сталь из ковша выливается в обычный сталеразливочный ковш, из которого уже поступает на МНЛЗ (машина непрерывного литья заготовок) или на разливку в слитки. Оттуда готовый продукт выходит в виде заготовок-слябов.

```python
def get_info_on_dataset(_df):
    
    CONST_IN_PERCENTAGES = float(100.0)
    
    try:
        _df.info()
        df_review_info = pd.DataFrame({'Not_Valid_Values_in %': (_df.isna().sum()/len(_df))*CONST_IN_PERCENTAGES} )
        print('')
        display(df_review_info.sort_values(by='Not_Valid_Values_in %', ascending=False).round(2))
        print('')
        print('Количество повторяющихся строк в датасете: {}'.format(_df.duplicated().sum()))
        print('')
        print('Первые 5 строк датасета:')
        display(_df.head(5))
        print('')
        print('Последние 5 строк датасета:')
        display(_df.tail(5))
        print('')
        print('Статистические данные:')
        display(_df.describe().round(2))
        
    except:
        print('Не могу предоставить отчет по датасету')
    
    pass


def plot_info_on_numerical_column( _df, 
                                   _feature_title,
                                   _label = False,
                                   _text_info = True,
                                   _bins = 50, 
                                   _figure_width = 17, 
                                   _figure_height = 7, 
                                   _fontsize=15,
                                   _number_of_decimals = 2,
                                   _title = ''):

    NCOLS_NUMBER_CONST = 2
        
    if _label == False:
        _label = _feature_title
      
    fig, ax = plt.subplots(ncols=NCOLS_NUMBER_CONST, figsize=(_figure_width, _figure_height))

    sns.histplot(data=_df, 
                 x=_feature_title, #
                 color="cornflowerblue", 
                 label=_label, #
                 bins=_bins, 
                 kde=True,
                 ax=ax[0])
        
    mean_value = _df[_feature_title].mean()  
    median_value = _df[_feature_title].median() 
    
    ax[0].axvline(x=mean_value, color='black', lw = 2, ls = ':', 
                  label = "mean: {}".format(round(mean_value, _number_of_decimals)))
    ax[0].axvline(x=median_value, color='blue', lw = 2, ls = '-.', 
                  label = "median: {}".format(round(median_value, _number_of_decimals)))

    ax[0].legend(fontsize = '15', title_fontsize = "10", loc='upper right') 
        
    plt.tight_layout()
    
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
    
    Q1 = _df[_feature_title].quantile(Q1_QUANTILE_VALUE_CONST)
    Q3 = _df[_feature_title].quantile(Q3_QUANTILE_VALUE_CONST)
    IQR = Q3 - Q1 

            
    outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
    outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR      
    
    sns.boxplot(_df[_feature_title], color="cornflowerblue", ax=ax[1])
    
    plt.suptitle(_title, fontsize=22, x=0.5, y=1.05)
    
    if _text_info:
        print("Q1 == {:.2f}, Q3 == {:.2f}, IQR == {:.2f}".format(Q1, Q3, IQR))
        print("[Q1 - 1.5*IQR] == {:.2f}, [Q3 + 1.5*IQR] == {:.2f}".format(outliers_low_bound, outliers_upper_bound))    

    plt.show()
    


def plot_df_coll_matrix(_df, 
                        _columns, 
                        _annotation = True,
                        _fiture_height = 15, 
                        _fiture_width = 15, 
                        _cell_annotation_font_size = 15,
                        _axiss_labels_font_size = 10):
 
    try:
        corr = _df[_columns].corr()
        fig, ax = plt.subplots(figsize=(_fiture_height, _fiture_width))           
        _pl = sns.heatmap(corr, 
                          ax=ax, 
                          annot=_annotation, 
                          linewidths=.5, 
                          fmt=".3f",
                          annot_kws={'size': _cell_annotation_font_size}, 
                          cmap='coolwarm',
                          square=True)
        
        _pl.set_yticklabels(ax.get_yticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
        _pl.set_xticklabels(ax.get_xticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
    except:
        print('Не могу построить график')    
        
    pass
```

```python
def plot_features_hists(_df,
                        _features_column_list,
                        _bins = 50, 
                        _figure_width = 20, 
                        _figure_height = 12, 
                        _colormaps_title = "Pastel2",
                        _stat = 'count',
                        _fontsize=15,
                        _title = ''):

    try:
        hists_number = len(_features_column_list)
        plt.figure(figsize=(_figure_width, _figure_height))
        hist_colors = get_cmap(_colormaps_title).colors 

        for _c, i in zip(_features_column_list, range(hists_number)):
            sns.histplot(data=_df, 
                         x=_c, 
                         color=hist_colors[i], 
                         label= str(_c), 
                         bins = _bins, 
                         stat=_stat,
                         kde=True)
        
        plt.ylabel('', fontsize=_fontsize)
        plt.xlabel('', fontsize=_fontsize)
        plt.yticks(fontsize=_fontsize)
        plt.xticks(fontsize=_fontsize)
        
        plt.suptitle(_title, fontsize=22, x=0.5, y=0.93)
        plt.legend(fontsize = '15', title_fontsize = "10")
        plt.show() 

    except:
        print('Не могу построить график')
        
    pass

def plot_features_boxplot(_df,
                          _figure_width = 12, 
                          _figure_height = 8,
                          _color = 'cornflowerblue',
                          _fontsize=15,
                          _title = ''):
  
    try:

        fig, ax = plt.subplots(figsize=(_figure_width, _figure_height))
        
        _pl = sns.boxplot(data=_df, 
                          orient="h", 
                          color=_color)
        
        _pl.set_yticklabels(ax.get_yticklabels(), size = _fontsize, fontweight = 'bold')
        #_pl.set_xticklabels(ax.get_xticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
        plt.suptitle(_title, fontsize=22, x=0.5, y=0.93)
        plt.show() 
        
    except:
        print('Не могу построить график')    
    pass
```

Как уже ранее мы обнаружили - в реактивной мощности есть большая отрицательная сумма.<br>
Посмотрим, что же такое активная и реактивная мощности:
_________________________________________
**Активная «полезная» мощность** — это та часть мощности, которая характеризует непосредственно процесс преобразования электрической энергии в некую другую энергию. Обозначается латинской буквой $P$ и измеряется в ваттах (Вт).

Рассчитывается по формуле: $P = U⋅I⋅cosφ$ ,
- где $U$ и $I$ – среднеквадратичное значение напряжения и силы тока цепи соответственно, $cos φ$ – косинус угла сдвига фазы между напряжением и током.


**Реактивная «вредная» мощность** — это мощность, которая образуется в процессе работы электроприборов с индуктивной или емкостной нагрузкой, и отражает происходящие электромагнитные колебания. Это энергия, которая переходит от источника питания к потребителю, а потом возвращается обратно в сеть.
Обозначается эта величина латинской буквой $Q$ и измеряется в вольт-амперах реактивных (Вар).

Рассчитывается по формуле: $Q = U⋅I⋅sinφ$,
- где $U$ и $I$ – среднеквадратичное значение напряжения и силы тока цепи соответственно, $sinφ$ – синус угла сдвига фазы между напряжением и током.

ВАЖНО! При расчете данная величина может быть как положительной, так и отрицательной – в зависимости от движения фазы.

**Полная мощность** – это геометрически вычисляемая величина, равная корню из суммы квадратов активной и реактивной мощностей соответственно. Обозначается латинской буквой $S$.


$ S = \sqrt{P^2 + Q^2} $,

где 
- $S$ - полная мощность,
- $P$ - активная мощность,
- $Q$ - реактивная мощность.
_________________________________________

То есть получается, что реактивная мощность может принимать и отрицательные значения. <br>

$ S = \sqrt{P^2 + Q^2} $,

где 
- $S$ - полная мощность,

```python
df_temp['q_measurements'] = df_temp.groupby('key')['key'].transform('count')
df_temp['time_measurement_first'] = df_temp.groupby('key')['time_measurement'].transform('min')
df_temp['time_measurement_last'] = df_temp.groupby('key')['time_measurement'].transform('max')
df_temp.head(10)
```

Сделаем суммарную таблицу по партиям, добавим колонки с первой и последней температурой и прошедшим между замерами временем.

```python
df_temp_fin = df_temp[(df_temp['time_measurement'] == df_temp['time_measurement_first']) 
                      | (df_temp['time_measurement'] == df_temp['time_measurement_last']
                        )].sort_values(by=['key','time_measurement'])

df_temp_fin['temp_first'] = np.where(df_temp_fin['time_measurement'] == df_temp_fin['time_measurement_first'],
                                     df_temp_fin['temperature'], 0)
df_temp_fin['temp_last'] = np.where(df_temp_fin['time_measurement'] == df_temp_fin['time_measurement_first'],
                                    df_temp_fin['temperature'].shift(-1), 0)

df_temp_fin = df_temp_fin.loc[df_temp_fin['temp_first'] > 0]
df_temp_fin = df_temp_fin.drop(['time_measurement','temperature'], axis=1)

df_temp_fin['time_minutes'] = round((df_temp_fin['time_measurement_last'] 
                                     - df_temp_fin['time_measurement_first']).dt.total_seconds()/60,2)

df_temp_fin.tail(10) 
```

```python
print('Cтатистические данные таблицы:')
df_temp_fin.describe().round(2)
```

## Созданение модели, подбор гиперпараметров

Перед нами стоит задача регрессии - предсказать конечную температуру партии стали. <br>
Целевая метрика **MAE** - cредняя абсолютная ошибка. <br> 
В MAE ошибка рассчитывается как среднее абсолютных разностей между целевыми значениями и прогнозами.


Возьмем для обучения 4 модели:
 - LinearRegression	
 - RandomForestRegressor	
 - XGBRegressor	
 - LGBMRegressor

Посмотрим какую метрику они нам дадут на подборе параметров, выберем модель с наименьшей ошибкой и протестируем ее. 


Создадим таблицу и функции для отображения и записи результатов обучения.

```python
results = pd.DataFrame(columns=['model_name',
                                'grid_search_time',
                                'mae_best_score', 
                                #'train_time', 
                                'mae_train', 
                                #'pred_time',
                                'mae_test'])

def ml_rmse(ml_target, ml_pred):
    return np.sqrt(mean_squared_error(ml_target, ml_pred))

def get_info_on_regression_results(row, _time, ml_target, ml_pred, _test = False):
    
    if _test==True: 
        #results.loc[row, 'pred_time'] = round(_time, 0)
        results.loc[row, 'mae_test'] = round(mean_absolute_error(ml_target, ml_pred),2)
    else:    
        #results.loc[row, 'train_time'] = round(_time, 0)
        results.loc[row, 'mae_train'] = round(mean_absolute_error(ml_target, ml_pred),2)

    print('MAE: ', round(mean_absolute_error(ml_target, ml_pred),2))
    print('RMSE:', round(ml_rmse(ml_target, ml_pred),2))
    print('Время операции {} сек.'.format(round(_time, 0) ))   
    
    pass

def get_results(row, model_name, _time, best_score, best_params):
    
    results.loc[row, 'model_name'] = model_name
    results.loc[row, 'grid_search_time'] = round(_time, 0)
    results.loc[row, 'mae_best_score'] = round(np.abs(gs_result.best_score_), 2)

    print('Время подбора {} сек.'.format(round(_time, 0) ))
    print('Лучший результат MAE: ', round(np.abs(gs_result.best_score_), 2))
    print('Лучшая модель с параметрами: ', gs_result.best_params_)
    
    pass
```

### LinearRegression

```python
lr_hyperparameters_search_space = dict()
lr_hyperparameters_search_space['n_jobs'] = [-1]
lr_hyperparameters_search_space['fit_intercept'] = [True,False]
```

```python
start = time.time()
print('Старт поиска лучших параметров для LR...')

model_lr = LinearRegression()
gs_lr = GridSearchCV(model_lr, 
                     lr_hyperparameters_search_space, 
                     scoring='neg_mean_absolute_error', 
                     n_jobs=-1, 
                     cv=CV_FOLDS_NUMBER_CONST)

gs_result = gs_lr.fit(X_train, Y_train)
print('Оптимальные параметры для LR найдены ...')

_time = time.time() - start
```

```python
get_results(1, 'LinearRegression', _time, gs_result.best_score_, gs_result.best_params_)
```

```python
start = time.time()
_dr_pred_median = DummyRegressor(strategy="median").fit(X_train, Y_train).predict(X_test)
_time = time.time() - start
mae_median = round(mean_absolute_error(Y_test, _dr_pred_median), 2)
print('Метрики для DummyRegressor стратегия "медиана"')
print('MAE: ', mae_median)

results.loc[7, 'model_name'] = 'DummyRegressor MEDIAN'
results.loc[7, 'grid_search_time'] = 0
results.loc[7, 'mae_best_score'] = mae_median
```
