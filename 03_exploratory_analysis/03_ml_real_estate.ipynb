```python
import warnings
import numpy as np
import pandas as pd
from pprint import pprint

import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot')

pd.set_option('display.max_rows', None)
pd.options.mode.chained_assignment = None
warnings.filterwarnings(action="ignore", message="Mean of empty slice")
```

```python
def get_info_on_dataset(_df, _full = True):
    """
    Get review on given dataset

    :param _df: pandas dataframe with data
    :param _full: True - provides full review, False - otherwise
    :return: None
    """ 
    CONST_IN_PERCENTAGES = float(100.0)
    
    try:
        _df.info()
        df_review_info = pd.DataFrame({'Not_Valid_Values_in %': (_df.isnull().sum()/len(_df))*CONST_IN_PERCENTAGES} )
        print("")
        print("Number of duplicated rows in the dataset: {}".format(_df.duplicated().sum()))
        print("")
        display(df_review_info.sort_values(by='Not_Valid_Values_in %', ascending=False))
    except:
        pprint("Can't provide review on given dataset")
    
    pass

def plot_df_coll_matrix(_df, _columns, _cell_annotation_font_size = 15):
    """ 
    Plot pairwise correlation matrix 
    :param _df_data: pandas dataframe with data
    :return: None 
    """   
    try:
        corr = _df[_columns].corr()
        fig, ax = plt.subplots(figsize=(10,8))           
        sns.heatmap(corr, annot=True, linewidths=.5, ax=ax, annot_kws={'size': _cell_annotation_font_size})
    except:
        pprint("Can't provide pairwise correlation plotting")
    
    pass
```

### Изучение параметров: площадь, цена, число комнат, высота потолков. Построение гистограмм для каждого параметра.

```python
def plot_hist(_x_data, 
              _bins = 100, 
              _x_ticks_range = [],
              _y_ticks_range = [],
              _outliers_removed = False,
              _kde = True,
              _x_axis_label = "x", 
              _y_axis_label = "count", 
              _width = 12, 
              _height = 8, 
              _fontsize=12):
    """ 
    Plot histogram
    :param _x_data: pandas dataframe series with x data
    :param _bins: bins number
    :param _outliers_removed: outliers_removed == True - plot _x_data with outliers removed, False - otherwise
    :param _kde: True -- kernel density estimation enabled, False --- otherwise
    :param _x_axis_label: x-axis label
    :param _y_axis_label: y-axis label    
    :param _width: width value
    :param _height: height value
    :param _fontsize: fontsize value
    :return: None 
    """   
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
        
    try:
        
        if(_outliers_removed):
            
            Q1 = _x_data.quantile(Q1_QUANTILE_VALUE_CONST)
            Q3 = _x_data.quantile(Q3_QUANTILE_VALUE_CONST)
            IQR = Q3 - Q1 
            
            outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
            outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR
            
            _x_data = _x_data[ (_x_data >= outliers_low_bound) & (_x_data <= outliers_upper_bound) ]
    
        plt.figure(figsize=(_width, _height))
        sns.histplot(x=_x_data, bins=_bins, kde=_kde)
        plt.xlabel(_x_axis_label, fontsize=_fontsize)
        plt.ylabel(_y_axis_label, fontsize=_fontsize)
        
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)
        
    except:
        pprint("Can't provide histogram plotting")
        
    pass

def plot_boxplot(_x_data, 
                 _bins = 100, 
                 _x_ticks_range = [],
                 _y_ticks_range = [],
                 _outliers_removed = False, 
                 _x_axis_label = "x", 
                 _y_axis_label = "count", 
                 _width = 8, 
                 _height = 6, 
                 _fontsize=12):
    """ 
    Plot histogram
    :param _x_data: pandas dataframe series with x data
    :param _bins: bins number
    :param _outliers_removed: outliers_removed == True - plot _x_data with outliers removed, False - otherwise
    :param _x_axis_label: x-axis label
    :param _y_axis_label: y-axis label    
    :param _width: width value
    :param _height: height value
    :param _fontsize: fontsize value
    :return: None 
    """   
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
        
    try:
        
        if(_outliers_removed):
            
            Q1 = _x_data.quantile(Q1_QUANTILE_VALUE_CONST)
            Q3 = _x_data.quantile(Q3_QUANTILE_VALUE_CONST)
            IQR = Q3 - Q1 
            
            outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
            outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR
            
            _x_data = _x_data[ (_x_data >= outliers_low_bound) & (_x_data <= outliers_upper_bound) ]

        plt.figure(figsize=(_width, _height))
                    
        sns.boxplot(x=_x_data, orient = 'h')
        plt.xlabel(_x_axis_label, fontsize=_fontsize)
        plt.ylabel(_y_axis_label, fontsize=_fontsize)
        
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)

    except:
        pprint("Can't provide boxplot plotting")
        
    pass

def get_box_and_whisker_info(_x_data):
    """ 
    Get info on box_and_whisker
    :param _x_data: pandas dataframe series with x data
    :return: None 
    """       
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
    
    try:
        Q1 = _x_data.quantile(Q1_QUANTILE_VALUE_CONST)
        Q3 = _x_data.quantile(Q3_QUANTILE_VALUE_CONST)
        IQR = Q3 - Q1 
            
        outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
        outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR      

        pprint("Q1 == {:.2f}, Q3 == {:.2f}, IQR == {:.2f}".format(Q1, Q3, IQR))
        pprint("[Q1 - 1.5*IQR] == {:.2f}, [Q3 + 1.5*IQR] == {:.2f}".format(outliers_low_bound, outliers_upper_bound))
        
    except:
        pprint("Can't provide box_and_whisker related info on given data")
        
    pass
```

```python
def plot_catplot(_df_data, 
                 _x_col_title,
                 _y_col_title,
                 _x_ticks_range = [],
                 _y_ticks_range = [],
                 _x_axis_label = "", 
                 _y_axis_label = "", 
                 _height= 1, 
                 _aspect = 1, 
                 _order = [],
                 _fontsize=12):
    """ 
    Plot catplot
    :param _df_data: pandas dataframe series with x data
    :param _x_axis_label: x-axis label
    :param _y_axis_label: y-axis label    
    :param _width: width value
    :param _height: height value
    :param _fontsize: fontsize value
    :return: None 
    """   

    try:
        
        #plt.figure(figsize=(_width, _height))
                    
        if _order:
            sns.catplot(x=_x_col_title, y=_y_col_title, kind="box", data=_df_data, height=_height, aspect=_aspect, palette='dark', order=_order)
        else:
            sns.catplot(x=_x_col_title, y=_y_col_title, kind="box", data=_df_data, height=_height, aspect=_aspect, palette='dark')
            
        #plt.xlabel(_x_axis_label, fontsize=_fontsize)
        #plt.ylabel(_y_axis_label, fontsize=_fontsize)
        
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)

    except:
        pprint("Can't provide catplot plotting")
        
    pass
```

## Вывод

После детального изучения предоставленной информации можно сказать, что в продаже преобладают 1, 2 и 3 комнатные квартиры, с общей площадью от 30 до 100 м2, с высотой полотков 2,65 м, диапазон цен объектов недвижимости от 1,2 М руб до 10 М руб.
В целом преобладают 1 и 2 комнатные квартиры, с площадью около 52 м2 по цене 4,6 М руб.

Самое большое влияние на стоимость объекта недвижимости оказывают удаленность от центра, общий метраж, площадь кухни и этаж, на котором находится объект. 
Чем ближе к центру, чем больше метраж - тем выше стоимость квартиры. При этом стоит отметить, что квартиры на первом этаже имеют самую низкую стоимость, далее по возрастанию стоимости следуют квартиры на последних этажах, а цены на не первый и не последний этаж самые высокие.
Другие факторы, как, например, наличие по близости парков, прудов и аэропортов, а также количество балконов влияют на стоимость квартиры, но не значительно.

Можно говорить, о том что квартира продана достаточно быстро, если объявление снято с публикации в пределах месяца, а долгие продажи квартиры - более 8-9 месяцев.
Что интересно на графике видно большие выбросы в районе публикации на 7, 30, 45, 60 и 90 дней - можно предположить, что это автоматические завершения публикаций по сроку давности без активности или без дальнейшей оплаты размещения таких объявлений.
 
Стоимость квартир самая высокая в радиусе 9 км от локального центра, а далее заметно снижается. 
Особо не обнаружено зависимости цены от дня публикации или месяца. Можно сказать, что наибольшая цена в апреле и сентябре, а в июне немного снижается. Возможно, колебания цены связаны с сезонностью, весной перед летним сезоном и осенью после летнего сезона цена немного выше, а летом наблюдается некоторое затишье на рынке недвижимости. 
Из предоставленной выборки самые высокие цены на квартиры в Ленинградской области наблюдаются в 2014 году, в 2016-2017 рынок заметно упал, но в 2019 видим небольшой рост цен.

Больше всего объявлений о продаже недвижимости, как и ожидалось, наблюдается в Санкт-Петербурге, наименьшее в Выборге. Из Топ10 максимальная цена за кв м в Санкт-Петербурге = 105 тыс руб м2, минимальная в Выборге = 58 тыс руб м2.

Все установленные ранее закономерности для объявлений в регионе справедливы и для объявлений в Санкт-Петербурге, с одной лишь, на мой взгляд, значимой разницей, что цены на квартиры выше всего в сентябре, могу допустить, что это связано с началом учебного года и концом отпусков.
 
Что интересно, в ходе проведенного анализа не обнаружена зависимость цены объекта от суммарного срока его публикации.
