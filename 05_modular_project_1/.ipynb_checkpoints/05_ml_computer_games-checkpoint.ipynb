```python
def get_info_on_dataset(_df, _full = True):
    """
    Get review on given dataset

    :param _df: pandas dataframe with data
    :param _full: True - provides full review, False - otherwise
    :return: None
    """ 
    CONST_IN_PERCENTAGES = float(100.0)
    
    try:
        _df.info()
        df_review_info = pd.DataFrame({'Not_Valid_Values_in %': (_df.isna().sum()/len(_df))*CONST_IN_PERCENTAGES} )
        print("")
        print("Number of duplicated rows in the dataset: {}".format(_df.duplicated().sum()))
        print("")
        display(df_review_info.sort_values(by='Not_Valid_Values_in %', ascending=False))
    except:
        pprint("Can't provide review on given dataset")
    
    pass

def plot_df_coll_matrix(_df, _columns, _cell_annotation_font_size = 15):
    """ 
    Plot pairwise correlation matrix 
    :param _df_data: pandas dataframe with data
    :return: None 
    """   
    try:
        corr = _df[_columns].corr()
        fig, ax = plt.subplots(figsize=(10,8))           
        sns.heatmap(corr, annot=True, linewidths=.5, ax=ax, cmap="Pastel2", annot_kws={'size': _cell_annotation_font_size})
    except:
        pprint("Can't provide pairwise correlation plotting")
    
    pass

def get_box_and_whisker_info(_x_data):
    """ 
    Get info on box_and_whisker
    :param _x_data: pandas dataframe series with x data
    :return: None 
    """       
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
    
    try:
        Q1 = _x_data.quantile(Q1_QUANTILE_VALUE_CONST)
        Q3 = _x_data.quantile(Q3_QUANTILE_VALUE_CONST)
        IQR = Q3 - Q1 
            
        outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
        outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR      

        pprint("Q1 == {:.2f}, Q3 == {:.2f}, IQR == {:.2f}".format(Q1, Q3, IQR))
        pprint("[Q1 - 1.5*IQR] == {:.2f}, [Q3 + 1.5*IQR] == {:.2f}".format(outliers_low_bound, outliers_upper_bound))
        
    except:
        pprint("Can't provide box_and_whisker related info on given data")
        
    pass
```

```python
def plot_hist(_x_data, 
              _bins = 100, 
              _x_ticks_range = [],
              _y_ticks_range = [],
              _outliers_removed = False,
              _kde = True,
              _x_axis_label = "x", 
              _y_axis_label = "count", 
              _width = 12, 
              _height = 8, 
              _fontsize=12):
    """ 
    Plot histogram
    :param _x_data: pandas dataframe series with x data
    :param _bins: bins number
    :param _outliers_removed: outliers_removed == True - plot _x_data with outliers removed, False - otherwise
    :param _kde: True -- kernel density estimation enabled, False --- otherwise
    :param _x_axis_label: x-axis label
    :param _y_axis_label: y-axis label    
    :param _width: width value
    :param _height: height value
    :param _fontsize: fontsize value
    :return: None 
    """   
    
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
        
    try:
        
        if(_outliers_removed):
            
            Q1 = _x_data.quantile(Q1_QUANTILE_VALUE_CONST)
            Q3 = _x_data.quantile(Q3_QUANTILE_VALUE_CONST)
            IQR = Q3 - Q1 
            
            outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
            outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR
            
            _x_data = _x_data[ (_x_data >= outliers_low_bound) & (_x_data <= outliers_upper_bound) ]
    
        fig = plt.figure(figsize=(_width, _height))
        sns.histplot(x=_x_data, bins=_bins, kde=_kde)
        plt.xlabel(_x_axis_label, fontsize=_fontsize)
        plt.ylabel(_y_axis_label, fontsize=_fontsize)
                
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)
        
    except:
        pprint("Can't provide histogram plotting")
        
    pass

def plot_boxplot(_x_data, 
                 _bins = 100, 
                 _x_ticks_range = [],
                 _y_ticks_range = [],
                 _outliers_removed = False, 
                 _x_axis_label = "x", 
                 _y_axis_label = "count", 
                 _width = 8, 
                 _height = 6, 
                 _fontsize=12):
    """ 
    Plot histogram
    :param _x_data: pandas dataframe series with x data
    :param _bins: bins number
    :param _outliers_removed: outliers_removed == True - plot _x_data with outliers removed, False - otherwise
    :param _x_axis_label: x-axis label
    :param _y_axis_label: y-axis label    
    :param _width: width value
    :param _height: height value
    :param _fontsize: fontsize value
    :return: None 
    """   
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
        
    try:
        
        if(_outliers_removed):
            
            Q1 = _x_data.quantile(Q1_QUANTILE_VALUE_CONST)
            Q3 = _x_data.quantile(Q3_QUANTILE_VALUE_CONST)
            IQR = Q3 - Q1 
            
            outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
            outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR
            
            _x_data = _x_data[ (_x_data >= outliers_low_bound) & (_x_data <= outliers_upper_bound) ]

        plt.figure(figsize=(_width, _height))
                    
        sns.boxplot(x=_x_data, orient = 'h', palette='pastel')
        plt.xlabel(_x_axis_label, fontsize=_fontsize)
        plt.ylabel(_y_axis_label, fontsize=_fontsize)
        
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)

    except:
        pprint("Can't provide boxplot plotting")
        
    pass

def plot_catplot(_df_data, 
                 _x_col_title,
                 _y_col_title,
                 _x_ticks_range = [],
                 _y_ticks_range = [],
                 _x_axis_label = "", 
                 _y_axis_label = "", 
                 _height= 1, 
                 _aspect = 1, 
                 _order = [],
                 _fontscale=2):
    """ 
    Plot catplot
    :param _df_data: pandas dataframe series with x data
    :param _x_axis_label: x-axis label
    :param _y_axis_label: y-axis label    
    :param _width: width value
    :param _height: height value
    :param _fontsize: fontsize value
    :return: None 
    """   

    try:
        
        sns.set(font_scale=_fontscale)
                    
        if _order:
            sns.catplot(x=_x_col_title, y=_y_col_title, kind="box", data=_df_data, height=_height, \
                        aspect=_aspect, palette='pastel', order=_order)
        else:
            sns.catplot(x=_x_col_title, y=_y_col_title, kind="box", data=_df_data, height=_height, \
                        aspect=_aspect, palette='pastel')
            
        sns.set()
        
        if _x_ticks_range:
            plt.xlim(*_x_ticks_range) 
        if _y_ticks_range:
            plt.ylim(*_y_ticks_range)

    except:
        pprint("Can't provide catplot plotting")
        
    pass
```

**Рейтинги**

- `EC`   Early childhood — Для детей младшего возраста: Игра подходит для детей от 3 лет и старше.
- `E`    Everyone — Для всех: Содержание вполне подходит для возрастной категории старше 6 лет.
- `E10+` Everyone 10 and older — Для всех старше 10 лет: Возрастная категория — от 10 лет.
- `T`    Teen — Подросткам 13—19 лет: Игра подходит для подростков старше 13 лет. 
- `M`    Mature — От 17 лет: Материалы игры не подходят для возрастной категории младше 17 лет. 
- `AO`   Adults Only 18+ — Только для взрослых: Содержание игры только для взрослых. 
- `RP`   Rating Pending — Рейтинг ожидается: Продукт был отправлен в ESRB и ожидает присвоение рейтинга. 

- `K-A`  Kids to Adults - устаревшая градация, перенесем его в Е.


 Формулируем нулевую и альтернативную гипотезы:
 - $H_0$ -- средние пользовательские рейтинги платформ Xbox One и PC --- одинаковы;
 - $H_1$ -- средние пользовательские рейтинги платформ Xbox One и PC отличаются друг от друга.
 
Критический уровень статистической значимости 5% ($\alpha = 0.05$)

Так как выборки независимые - воспользуемся функцией `ttest_ind`

Проверим гипотезу сначала для всего периода выборки:

```python
ALPHA_CONST_VALUE = float(.05)
VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST = float(0.5)
VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST = float(2.0)

_equal_var = False

xone_user_score_var = df_games.loc[ df_games['platform'] == "XOne", "users_score"].var()
pc_user_score_var = df_games.loc[ df_games['platform'] == "PC", "users_score"].var()

vars_ratio = xone_user_score_var/pc_user_score_var

print("Variances ratio equals to: {}".format(round(vars_ratio,3)))
try:
    if  (vars_ratio > VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST) and (vars_ratio < VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST):
        _equal_var = True
        print("Variances are similar")
    else:
        print("Variances are not similar")        
except:
    pass

p_value = st.ttest_ind(df_games.loc[ (df_games['platform'] == "XOne") & (~df_games['users_score'].isna()), "users_score"],
                       df_games.loc[ (df_games['platform'] == "PC") & ~df_games['users_score'].isna(), "users_score"], 
                       equal_var = _equal_var).pvalue

print(" ")
if p_value < ALPHA_CONST_VALUE:
    print("p_value < ALPHA_CONST_VALUE ==> нулевая гипотеза отвергается.")
else:
    print("p_value > ALPHA_CONST_VALUE ==> нулевая гипотеза не может быть отвергнута.")

print("p_value == {}, alpha = {}.".format(p_value, ALPHA_CONST_VALUE))

mean_xone_value = df_games.loc[ (df_games['platform'] == "XOne") & (~df_games['users_score'].isna()), "users_score"].mean()   
mean_pc_value = df_games.loc[ (df_games['platform'] == "PC") & ~df_games['users_score'].isna(), "users_score"].mean() 

print(" ")
print("Средний пользовательский рейтинг платформы Xbox One: {}".format(round(mean_xone_value,2)))
print("Средний пользовательский рейтинг платформы PC: {}".format(round(mean_pc_value,2)))
```

А теперь проверим гипотезу для актуального периода выборки с 2012 по 2016 годы:

```python
ALPHA_CONST_VALUE = float(.05)
VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST = float(0.5)
VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST = float(2.0)

_equal_var = False

xone_user_score_var = df_games_actual.loc[ df_games['platform'] == "XOne", "users_score"].var()
pc_user_score_var = df_games_actual.loc[ df_games['platform'] == "PC", "users_score"].var()

vars_ratio = xone_user_score_var/pc_user_score_var

print("Variances ratio equals to: {}".format(round(vars_ratio,3)))
try:
    if  (vars_ratio > VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST) and (vars_ratio < VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST):
        _equal_var = True
        print("Variances are similar")
    else:
        print("Variances are not similar")        
except:
    pass

p_value = st.ttest_ind(df_games_actual.loc[ (df_games_actual['platform'] == "XOne") & \
                                           (~df_games_actual['users_score'].isna()), "users_score"],
                       df_games_actual.loc[ (df_games_actual['platform'] == "PC") & \
                                           ~df_games_actual['users_score'].isna(), "users_score"], 
                       equal_var = _equal_var).pvalue

print(" ")
if p_value < ALPHA_CONST_VALUE:
    print("p_value < ALPHA_CONST_VALUE ==> нулевая гипотеза отвергается.")
else:
    print("p_value > ALPHA_CONST_VALUE ==> нулевая гипотеза не может быть отвергнута.")

print("p_value == {}, alpha = {}.".format(p_value, ALPHA_CONST_VALUE))

mean_xone_value = df_games_actual.loc[ (df_games_actual['platform'] == "XOne") & \
                                      (~df_games_actual['users_score'].isna()), "users_score"].mean()   
mean_pc_value = df_games_actual.loc[ (df_games_actual['platform'] == "PC") & \
                                    ~df_games_actual['users_score'].isna(), "users_score"].mean() 

print(" ")
print("Средний пользовательский рейтинг платформы Xbox One: {}".format(round(mean_xone_value,2)))
print("Средний пользовательский рейтинг платформы PC: {}".format(round(mean_pc_value,2)))
```

Критический уровень статистической значимости 5% ($\alpha = 0.05$)

Проверим гипотезу сначала для всего периода выборки:

```python
ALPHA_CONST_VALUE = float(.05)
VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST = float(0.5)
VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST = float(2.0)

_equal_var = False

action_user_score_var = df_games.loc[ df_games['genre'] == "Action", "users_score"].var()
sports_user_score_var = df_games.loc[ df_games['genre'] == "Sports", "users_score"].var()

vars_ratio = action_user_score_var/sports_user_score_var

print("Variances ratio equals to: {}".format(round(vars_ratio,3)))
try:
    if  (vars_ratio > VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST) and (vars_ratio < VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST):
        _equal_var = True
        print("Variances are similar.")
    else:
        print("Variances are not similar")        
except:
    pass

p_value = st.ttest_ind(df_games.loc[ (df_games['genre'] == "Action") & (~df_games['users_score'].isna()), "users_score"],
                       df_games.loc[ (df_games['genre'] == "Sports") & ~df_games['users_score'].isna(), "users_score"], 
                       equal_var = _equal_var).pvalue

print(" ")
if p_value < ALPHA_CONST_VALUE:
    print("p_value < ALPHA_CONST_VALUE ==> нулевая гипотеза отвергается.")
else:
    print("p_value > ALPHA_CONST_VALUE ==> нулевая гипотеза не может быть отвергнута.")

print("p_value == {}, alpha = {}.".format(p_value, ALPHA_CONST_VALUE))

mean_action_value = df_games.loc[ (df_games['genre'] == "Action") & (~df_games['users_score'].isna()), "users_score"].mean()   
mean_sports_value = df_games.loc[ (df_games['genre'] == "Sports") & ~df_games['users_score'].isna(), "users_score"].mean() 

print(" ")
print("Средний пользовательский рейтинг жанра Action: {}".format(round(mean_action_value,2)))
print("Средний пользовательский рейтинг жанра Sports: {}".format(round(mean_sports_value,2)))
```

А теперь проверим гипотезу для актуального периода выборки с 2012 по 2016 годы:

```python
ALPHA_CONST_VALUE = float(.05)
VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST = float(0.5)
VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST = float(2.0)

_equal_var = False

action_user_score_var = df_games_actual.loc[ df_games_actual['genre'] == "Action", "users_score"].var()
sports_user_score_var = df_games_actual.loc[ df_games_actual['genre'] == "Sports", "users_score"].var()

vars_ratio = action_user_score_var/sports_user_score_var

print("Variances ratio equals to: {}".format(round(vars_ratio,3)))
try:
    if  (vars_ratio > VARIANCE_RULE_OF_THUMB_LEFT_BOUND_CONST) and (vars_ratio < VARIANCE_RULE_OF_THUMB_RIGHT_BOUND_CONST):
        _equal_var = True
        print("Variances are similar.")
    else:
        print("Variances are not similar")        
except:
    pass

p_value = st.ttest_ind(df_games_actual.loc[ (df_games_actual['genre'] == "Action") & \
                                           (~df_games_actual['users_score'].isna()), "users_score"],
                       df_games_actual.loc[ (df_games_actual['genre'] == "Sports") & \
                                           ~df_games_actual['users_score'].isna(), "users_score"], 
                       equal_var = _equal_var).pvalue

print(" ")
if p_value < ALPHA_CONST_VALUE:
    print("p_value < ALPHA_CONST_VALUE ==> нулевая гипотеза отвергается.")
else:
    print("p_value > ALPHA_CONST_VALUE ==> нулевая гипотеза не может быть отвергнута.")

print("p_value == {}, alpha = {}.".format(p_value, ALPHA_CONST_VALUE))

mean_action_value = df_games_actual.loc[ (df_games_actual['genre'] == "Action") & \
                                        (~df_games_actual['users_score'].isna()), "users_score"].mean()   
mean_sports_value = df_games_actual.loc[ (df_games_actual['genre'] == "Sports") & \
                                        ~df_games_actual['users_score'].isna(), "users_score"].mean() 

print(" ")
print("Средний пользовательский рейтинг жанра Action: {}".format(round(mean_action_value,2)))
print("Средний пользовательский рейтинг жанра Sports: {}".format(round(mean_sports_value,2)))
```
