# Описание проекта

Для добывающей компании **«ГлавРосГосНефть»** нужно решить, где бурить новую скважину.

Шаги для выбора локации обычно такие:
- В избранном регионе собирают характеристики для скважин: качество нефти и объём её запасов;
- Строят модель для предсказания объёма запасов в новых скважинах;
- Выбирают скважины с самыми высокими оценками значений;
- Определяют регион с максимальной суммарной прибылью отобранных скважин.

Нам предоставлены пробы нефти в трёх регионах. <br>
Характеристики для каждой скважины в регионе уже известны.<br>
Построим модель для определения региона, где добыча принесёт наибольшую прибыль.<br>
Проанализируем возможную прибыль и риски техникой Bootstrap.

**Шаги нашего исследования**
- Загрузим и подготовим данные.
- Обучим и проверим модель для каждого региона.
- Подготовимся к расчету прибыли.
- Рассчитаем прибыль по выбранным скважинам и предсказаниям модели.
- Посчитаем риски и прибыль для каждого региона.


**Из условий задачи:**<br>
Для обучения модели подходит только линейная регрессия.<br>
При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.<br>
Бюджет на разработку скважин в регионе — 10 млрд рублей.<br>
При нынешних ценах один баррель сырья приносит 450 рублей дохода.<br>
Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.<br>
После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%.<br>
Среди них выбирают регион с наибольшей средней прибылью.<br>

Данные синтетические: детали контрактов и характеристики месторождений не разглашаются.

```python
def get_info_on_dataset(_df, _full = False):
    """
    Get review on given dataset

    :param _df: pandas dataframe with data
    :param _full: True - provides full review, False - otherwise
    :return: None
    """ 
    CONST_IN_PERCENTAGES = float(100.0)
    
    try:
        _df.info()
        df_review_info = pd.DataFrame({'Not_Valid_Values_in %': (_df.isna().sum()/len(_df))*CONST_IN_PERCENTAGES} )
        print("")
        display(df_review_info.sort_values(by='Not_Valid_Values_in %', ascending=False))
     
        if(_full):
            _s_nunique = _df.nunique()
            _df_nunique = pd.DataFrame({'DataFrame Columns':_s_nunique.index, 
                                        'Unique Values Number':_s_nunique.values})
            display(_df_nunique)
        
        print("")
        print("Number of duplicated rows in the dataset: {}".format(_df.duplicated().sum()))
            
    except:
        pprint("Can't provide review on given dataset")
    
    pass
```

```python
def plot_df_coll_matrix(_df, 
                        _columns,
                        _name,
                        _fiture__height = 15, 
                        _fiture_width = 15, 
                        _cell_annotation_font_size = 15,
                        _axiss_labels_font_size = 10):
    """ 
    Plot pairwise correlation matrix 
    :param _df_data: pandas dataframe with data
    :return: None 
    """   
    try:
        corr = _df[_columns].corr()
        fig, ax = plt.subplots(figsize=(_fiture__height, _fiture_width))           
        _pl = sns.heatmap(corr, 
                          ax=ax, 
                          annot=True, 
                          linewidths=.5, 
                          fmt=".3f",
                          annot_kws={'size': _cell_annotation_font_size}, 
                          cmap='coolwarm',
                          square=True)
        
        _pl.set_yticklabels(ax.get_yticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
        _pl.set_xticklabels(ax.get_xticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
        _pl.title.set_text(_name)
        
    except:
        pprint("Can't provide pairwise correlation plotting")    
    pass
```

```python
def plot_info_on_numerical_column( _df, 
                                   _feature_title,
                                   _text_info = True,
                                   _bins = 50, 
                                   _figure_width = 20, 
                                   _figure_height = 12, 
                                   _fontsize=15,
                                   _number_of_decimals = 2):
    """ 
    Plot info on numerical feature
    param;  _df                    : pandas dataframe
    param:  _feature_title         : feature's column title
    param:  _bins:                 : hists bins number  
    param:  _fiture__width         : figure's width value
    param:  _fiture_height         : fiture's height value
    param:  _fontsize              : fontsize value
    :return                        : None 
    """     

    NCOLS_NUMBER_CONST = 2
    
    fig, ax = plt.subplots(ncols=NCOLS_NUMBER_CONST, figsize=(_figure_width, _figure_height))

    sns.histplot(data=_df, 
                 x=_feature_title, 
                 color="steelblue", 
                 label=_feature_title, 
                 bins=_bins, 
                 kde=True,
                 ax=ax[0])
        
    mean_value = _df[_feature_title].mean()  
    median_value = _df[_feature_title].median() 
    
    ax[0].axvline(x=mean_value, 
                  color='black', 
                  lw = 2, 
                  ls = ':', 
                  label = "mean: {}".format(round(mean_value, _number_of_decimals)))
    ax[0].axvline(x=median_value, 
                  color='blue', 
                  lw = 2, 
                  ls = '-.', 
                  label = "median: {}".format(round(median_value, _number_of_decimals)))

    ax[0].legend(fontsize = '15', title_fontsize = "10", loc='upper right') 
        
    plt.tight_layout()
    
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
    
    Q1 = _df[_feature_title].quantile(Q1_QUANTILE_VALUE_CONST)
    Q3 = _df[_feature_title].quantile(Q3_QUANTILE_VALUE_CONST)
    IQR = Q3 - Q1 

            
    outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
    outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR      
    
    sns.boxplot(_df[_feature_title], color="steelblue", ax=ax[1])

    if _text_info:
        print("Q1 == {:.2f}, Q3 == {:.2f}, IQR == {:.2f}".format(Q1, Q3, IQR))
        print("[Q1 - 1.5*IQR] == {:.2f}, [Q3 + 1.5*IQR] == {:.2f}".format(outliers_low_bound, outliers_upper_bound))    

    plt.show()
```

## Создание и обучение модели.

Разобьем данные на обучающую и валидационную выборки в соотношении 75:25. <br>
Создадим и обучим модель LinearRegression, подберем лучшие гиперпараметры, сохраним предсказания и правильные ответы на валидационной выборке.

```python
def get_lr_prediction(_df_list, _ml_target_column_title):
    
    lr_prediction_results = []
    
    lr_hyperparameters_search_space = dict()
    lr_hyperparameters_search_space['n_jobs'] = [-1]
    lr_hyperparameters_search_space['normalize'] = [True,False]
    lr_hyperparameters_search_space['fit_intercept'] = [True,False]  

    for _i, _df in enumerate(_df_list):

        X = _df.drop([_ml_target_column_title], axis=1).values   
        y = _df[_ml_target_column_title].values
        
        
        X_training, X_validation, y_training, y_validation = train_test_split(X, y, 
                                                                  test_size=ML_VALIDATION_SAMPLES_SIZE_IN_PERCENT_CONST, 
                                                                  random_state=ML_RND_GNR_SEED_VALUE_CONST)

        print("Вычисления для датасета: {}...".format(_i))
        print("Обучающая выборка:")
        print("Размерность X: {}, размерность y: {}".format(X_training.shape, y_training.shape))
        print("Валидационная выборка:")
        print("Размерность X: {}, размерность y: {}".format(X_validation.shape, y_validation.shape))

        start = time.time()
        print("Обучение модели началось...")
        
        gs_lr = GridSearchCV(LinearRegression(), 
                             lr_hyperparameters_search_space, 
                             scoring='r2',
                             cv=CROSS_VALIDATION_FOLDS_NUMBER_CONST)
        
        gs_lr.fit(X_training, y_training)
        print('Лучший результат по критерию r2: %s' % round(gs_lr.best_score_, 3))
        print('Значения оптимальных параметров для лучшей модели: %s' % gs_lr.best_params_)
        
        _rs_pred = gs_lr.predict(X_validation)
        
        _df_data_dict = {'actual_well_reserves':y_validation, 'predicted_well_reserves':_rs_pred, 'location':_i}
        
        end = time.time()
        print("Продолжительность вычислений {} sec.".format( round(end - start, 2) ))
        print("")
                
        lr_prediction_results.append(pd.DataFrame.from_dict(_df_data_dict))
    
    return pd.concat(lr_prediction_results)
```

```python
df_geo_data_list = [df_geo_data_0, df_geo_data_1, df_geo_data_2]

df_total_results = get_lr_prediction(df_geo_data_list, ML_REGRESSION_PROBLEM_TARGET_CONST)
```

Рассчитаем и добавим к нашей таблице метрики RMSE, MSE, R2

```python
def compute_dfs_metric(_df, 
                       actual_well_reserves_column,
                       predicted_well_reserves_column,
                       location_well_reserved_column,
                       _metric_f):

    RMSE_METRIC_FUNCTION_TITLE_CONST = 'root_mean_squared_error'
    
    _metric_res = []    
    _locations_list = _df.location.unique().tolist()
    
    try:
        
        for i in _locations_list:
            
            _df_i = _df.query("{} == {}".format(location_well_reserved_column, i))
            
            _actual_well_reserves = _df_i[actual_well_reserves_column]
            _predicted_well_reserves = _df_i[predicted_well_reserves_column]

            if _metric_f == RMSE_METRIC_FUNCTION_TITLE_CONST:           
                _actual_metric_value = np.sqrt(mean_squared_error(_actual_well_reserves,_predicted_well_reserves))        
            else:
                _actual_metric_value = _metric_f(_actual_well_reserves, _predicted_well_reserves)             

            _metric_res.append(_actual_metric_value)            
    except:
        
        print("Can't provide computation for the given dataset & metric")
    
    return _metric_res
```

```python
df_wells_location_info = df_total_results.pivot_table(index='location', aggfunc=['mean', 'median'])

df_wells_location_info['r2'] = compute_dfs_metric(_df = df_total_results, 
                                                    actual_well_reserves_column = 'actual_well_reserves', 
                                                    predicted_well_reserves_column = 'predicted_well_reserves', 
                                                    location_well_reserved_column = 'location',
                                                    _metric_f=r2_score)

df_wells_location_info['mae'] = compute_dfs_metric(_df = df_total_results, 
                                                    actual_well_reserves_column = 'actual_well_reserves', 
                                                    predicted_well_reserves_column = 'predicted_well_reserves', 
                                                    location_well_reserved_column = 'location',
                                                    _metric_f=mean_absolute_error)

df_wells_location_info['rmse'] = compute_dfs_metric(_df = df_total_results, 
                                                    actual_well_reserves_column = 'actual_well_reserves', 
                                                    predicted_well_reserves_column = 'predicted_well_reserves', 
                                                    location_well_reserved_column = 'location',
                                                    _metric_f='root_mean_squared_error')

display(df_wells_location_info.T)
```

**Вывод**

Видим, что значения прогнозных средней и медианы близки к фактическим значениям средней и медианы по регионам.

Если оценивать метрики, то:
- по значениям RMSE и MSE наиболее точная оказалось модель региона с датасетом df_geo_data_1 (чем меньше эти метрики, тем лучше),
- по значению R2 так же лучше модель для региона df_geo_data_1, данная метрика близка к 1 (это говорит, о том, что модель нашей линейной регрессии практически идеальна).

Но не будем забывать, что данный датасет df_geo_data_1 показал высокую корреляцию столбцов `f_2` и `well_reserves`, и также столбец  `well_reserves` содержит всего 12 уникальных значений  - это, я думаю, объясняет высокие результаты модели для данного региона.

## Подготовка к расчёту прибыли

Рассчитаем достаточный объём сырья для безубыточной разработки новой скважины и сравним полученный объём сырья со средним запасом в каждом регионе.

```python
OPTIONAL_WELLS_NUMBER_CONST = 500
EXTRACTION_WELLS_NUMBER_CONST = 200

TOTAL_BUDGET_VALUE_CONST = float(10e9)
ONE_THOUSAND_BARRELS_REVENUE_CONST = float(45e4)
```

## Расчёт прибыли и рисков 

Далее выберем 200 скважин с максимальными значениями предсказаний, просуммируем целевое значение объёма сырья, соответствующее этим предсказаниям. И рассчитаем прибыль для полученного объёма сырья.

## Риски и прибыль для каждого региона


Применим Bootstrap с 1000 выборок для поиска распределения прибыли. <br>
Найдем среднюю прибыль, 95%-й доверительный интервал и риск убытков. <br>

```python
df_total_results['actual_well_revenue'] = df_total_results['actual_well_reserves']*ONE_THOUSAND_BARRELS_REVENUE_CONST
df_total_results.head(5)
```

```python
def provide_bootstrap_revenue_simulation(_df, _total_experiments_number=1000):
    
    _random_state = np.random.RandomState(ML_RND_GNR_SEED_VALUE_CONST)
    revenue_values = []
    
    for i in range(_total_experiments_number):
        
        _df_sample = _df.sample(n=OPTIONAL_WELLS_NUMBER_CONST, 
                                random_state=_random_state, 
                                replace=True)
        
        
        df_best_n_wells = _df_sample.sort_values(by='predicted_well_reserves', ascending=False)\
                                                                               [:EXTRACTION_WELLS_NUMBER_CONST]
        actual_well_profit = df_best_n_wells['actual_well_revenue'].sum() - TOTAL_BUDGET_VALUE_CONST
        revenue_values.append(actual_well_profit)    
        
    return revenue_values    
```

```python
def provide_info_on_revenue_simulation(_revenue_simulation_values):
    
  
    _revenue_simulation_values_s = pd.Series(_revenue_simulation_values)
    
    _revenue_simulation_values_s = _revenue_simulation_values_s*BILLION_COEFF_CONST
    
    quantile_0_025 = _revenue_simulation_values_s.quantile(0.025)
    quantile_0_975 = _revenue_simulation_values_s.quantile(0.975)
    
    loss_of_revenue = _revenue_simulation_values_s[_revenue_simulation_values_s<.0]
    loss_of_revenue_probability = float(loss_of_revenue.count())/float(len(_revenue_simulation_values))
    
    revenue_mean = _revenue_simulation_values_s.mean()

    
    print("Средняя прибыль: {} млрд.руб.".format(round(revenue_mean,3)))
    print("Доверительный интервал: ({}, {})".format(round(quantile_0_025,3), round(quantile_0_975,3)))
    print("Вероятность события 'отрицательный доход': {}".format(loss_of_revenue_probability))
    print("")
    df_simulated_revenue = pd.DataFrame({'simulated_revenue': _revenue_simulation_values})
    df_simulated_revenue['simulated_revenue [in bil]'] = df_simulated_revenue['simulated_revenue']*BILLION_COEFF_CONST
    
    plot_info_on_numerical_column(df_simulated_revenue, 
                                  'simulated_revenue [in bil]', 
                                  _figure_width = 16, 
                                  _figure_height = 6,
                                  _number_of_decimals = 3)
```

```python
revenue_values_l0 = provide_bootstrap_revenue_simulation(df_total_results.query("location == 0"))
revenue_values_l1 = provide_bootstrap_revenue_simulation(df_total_results.query("location == 1"))
revenue_values_l2 = provide_bootstrap_revenue_simulation(df_total_results.query("location == 2"))
```

```python
provide_info_on_revenue_simulation(revenue_values_l0)
provide_info_on_revenue_simulation(revenue_values_l1)
provide_info_on_revenue_simulation(revenue_values_l2)
```

**Вывод** <br>
При помощи техники Bootstrap с 1000 выборок мы построили распределение прибыли, нашли среднюю прибыль, 95% доверительный интервал и риск убытков.<br>
Наибольшая средняя прибыль в регионе **df_geo_data_1** = 0.488 млрд рублей, также в этом регионе наименьшая вероятность отрицательного дохода = 0.01.<br>
Менее привлекателен регион **df_geo_data_0**: средняя прибыль = 0.461 млрд рублей, вероятность отрицательного дохода = 0.038.<br>
Самый рискованный регион **df_geo_data_2**: средняя прибыль = 0.379 млрд рублей, вероятность отрицательного дохода = 0.082.
