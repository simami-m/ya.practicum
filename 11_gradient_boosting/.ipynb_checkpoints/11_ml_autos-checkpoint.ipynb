```python
def get_info_on_dataset(_df):
    
    CONST_IN_PERCENTAGES = float(100.0)
    
    try:
        _df.info()
        df_review_info = pd.DataFrame({'Not_Valid_Values_in %': (_df.isna().sum()/len(_df))*CONST_IN_PERCENTAGES} )
        print("")
        display(df_review_info.sort_values(by='Not_Valid_Values_in %', ascending=False))
            
        print("")
        print("Number of duplicated rows in the dataset: {}".format(_df.duplicated().sum()))
            
    except:
        print("Can't provide review on given dataset")
    
    pass
```

```python
def plot_info_on_numerical_column( _df, 
                                   _feature_title,
                                   _text_info = True,
                                   _bins = 50, 
                                   _figure_width = 20, 
                                   _figure_height = 12, 
                                   _fontsize=15,
                                   _number_of_decimals = 2):
    
    NCOLS_NUMBER_CONST = 2
    
    fig, ax = plt.subplots(ncols=NCOLS_NUMBER_CONST, figsize=(_figure_width, _figure_height))

    sns.histplot(data=_df, 
                 x=_feature_title, 
                 color="steelblue", 
                 label=_feature_title, 
                 bins=_bins, 
                 kde=True,
                 ax=ax[0])
        
    mean_value = _df[_feature_title].mean()  
    median_value = _df[_feature_title].median() 
    
    ax[0].axvline(x=mean_value, color='black', lw = 2, ls = ':', label = "mean: {}".format(
        round(mean_value, _number_of_decimals)))
    ax[0].axvline(x=median_value, color='blue', lw = 2, ls = '-.', label = "median: {}".format(
        round(median_value, _number_of_decimals)))

    ax[0].legend(fontsize = '15', title_fontsize = "10", loc='upper right') 
        
    plt.tight_layout()
    
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
    
    Q1 = _df[_feature_title].quantile(Q1_QUANTILE_VALUE_CONST)
    Q3 = _df[_feature_title].quantile(Q3_QUANTILE_VALUE_CONST)
    IQR = Q3 - Q1 

            
    outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
    outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR      
    
    sns.boxplot(_df[_feature_title], color="steelblue", ax=ax[1])

    if _text_info:
        print("Q1 == {:.2f}, Q3 == {:.2f}, IQR == {:.2f}".format(Q1, Q3, IQR))
        print("[Q1 - 1.5*IQR] == {:.2f}, [Q3 + 1.5*IQR] == {:.2f}".format(outliers_low_bound, outliers_upper_bound))    

    plt.show()
```

```python
def plot_df_coll_matrix(_df, 
                        _columns, 
                        _annotation = True,
                        _figure_height = 15, 
                        _figure_width = 15, 
                        _cell_annotation_font_size = 15,
                        _axiss_labels_font_size = 10):

    try:
        corr = _df[_columns].corr()
        fig, ax = plt.subplots(figsize=(_figure_height, _figure_width))           
        _pl = sns.heatmap(corr, 
                          ax=ax, 
                          annot=_annotation, 
                          linewidths=.5, 
                          fmt=".3f",
                          annot_kws={'size': _cell_annotation_font_size}, 
                          cmap='coolwarm',
                          square=True)
        
        _pl.set_yticklabels(ax.get_yticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
        _pl.set_xticklabels(ax.get_xticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
    except:
        print("Can't provide pairwise correlation plotting")    
    pass
```

```python
def plot_features_hist(_df,
                       _x_feature_column,
                       _hue_feature_column,
                       _bins = 50, 
                       _figure_width = 12, 
                       _figure_height = 20, 
                       _palette = 'Set2',
                       _fontsize=15,
                       _axiss_labels_font_size = 15):
   
    try:
        
        fig, ax = plt.subplots(figsize=(_figure_height, _figure_width)) 
        
        _pl = sns.histplot(data=df_autos, 
                           x=_x_feature_column, 
                           hue=_hue_feature_column, 
                           kde=True, 
                           bins=_bins, 
                           palette=_palette);
        
        _pl.set_yticklabels(_pl.get_yticks(), size = _axiss_labels_font_size)
        _pl.set_xticklabels(_pl.get_xticks(), size = _axiss_labels_font_size)

        _pl.set_xlabel(_pl.get_xlabel(),fontsize=_fontsize, fontweight = 'bold')
        _pl.set_ylabel(_pl.get_ylabel(),fontsize=_fontsize, fontweight = 'bold')

        plt.setp(ax.get_legend().get_texts(), fontsize=_fontsize) 
        plt.setp(ax.get_legend().get_title(), fontsize=_fontsize) 

        plt.show()    
        
    except:
        print("Can't provide boxplot plotting")
        
    pass
```

```python
def plot_features_importances(_df,
                              _figure_width = 17, 
                              _figure_height = 10, 
                              _fontsize=15):
    
    fig, ax = plt.subplots(figsize=(_figure_width, _figure_height)) 

    _pl = sns.barplot(x="importances_value", 
                      y="features", 
                      data=_df, 
                      orient="h", 
                      order=_df.sort_values(by="importances_value", ascending=False).features.tolist(), 
                      color="c");

    _pl.set_yticklabels(_pl.get_yticklabels(), size = _fontsize)
    _pl.set_xticklabels(_pl.get_xticks(), size = _fontsize)

    _pl.set_xlabel(_pl.get_xlabel(),fontsize=_fontsize, fontweight = 'bold')
    _pl.set_ylabel(_pl.get_ylabel(),fontsize=_fontsize, fontweight = 'bold')
    
    pass
```

### Подготовка данных для моделей

#### Кодирование категориальных признаков

Подготовим наши категориальные признаки для работы с моделями машинного обучения. <br>
К столбцам с небольшим количеством уникальных значений применим кодирование One Hot Encoder, <br>
для остальных (с большим количеством категорий) применим Ordinal Encoder.

```python
all_ml_columns_list = ['kilometer',
                       'fuel_type',
                       'power',
                       'brand', 
                       'vehicle_type',
                       'gearbox',
                       'model',   
                       'not_repaired'
                       'registration_year', 
                       'registration_month',
                       'price'
]

one_hot_encoder_columns_list = ['fuel_type',
                                'gearbox',
                                'not_repaired'
]

ordinal_encoder_columns_list = ['brand', 
                                'vehicle_type',
                                'model']


numerical_features_columns_list = ['kilometer',
                                   'power',
                                   'registration_year', 
                                   'registration_month'
]
```

```python
df_autos_cp = df_autos.copy()
df_autos_features = pd.get_dummies(df_autos_cp[one_hot_encoder_columns_list], drop_first=False)
df_autos_features.head(3).T
```

```python
encoder = OrdinalEncoder()
encoder.fit(df_autos_cp[ordinal_encoder_columns_list])

df_autos_features[ordinal_encoder_columns_list] = encoder.transform(df_autos_cp[ordinal_encoder_columns_list])

df_autos_features.head(10).T
```

Соберем полную таблицу с обучающими признаками.

```python
df_autos_features[numerical_features_columns_list] = df_autos[numerical_features_columns_list]
df_autos_features.head(10).T
```

#### Разделение на выборки и масшабирование

Создаем массивы NumPy с целевым и обучающими признаками.

```python
X_main = df_autos_features.to_numpy()
Y_main = df_autos['price'].to_numpy().astype(float)
```

```python
print('Размерность Массива NumPy с обучающиими признаками: ', X_main.shape)
print('Размерность Массива NumPy с целевым признаком:    ', Y_main.shape)
print()
print('Размерность датасета с обучающиими признаками:      ', df_autos_features.shape)
print('Размерность датасета с целевым признаком:         ', df_autos.price.shape)
```

Разобъем на обучающую и тестовые выборки в соотношении 70/30

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_main, 
                                                    Y_main, 
                                                    test_size=0.3, 
                                                    random_state=ML_RND_GNR_SEED_VALUE_CONST)

print('\nОбучающая выборка обучающие признаки:')
print('Размерность: {}'.format(X_train.shape))
print('\nОбучающая выборка целевые признаки:')
print('Размерность: {}'.format(Y_train.shape))
print('\nТестовая выборка обучающие признаки:')
print('Размерность: {}'.format(X_test.shape))
print('\nТестовая выборка целевые признаки:')
print('Размерность: {}'.format(Y_test.shape))

print('\nОбучающая выборка обучающие признаки:')
display(X_train[:1])
print('\nОбучающая выборка целевые признаки:')
display(Y_train[:])
print('\nТестовая выборка обучающие признаки:')
display(X_test[:1])
print('\nТестовая выборка целевые признаки:')
display(Y_test[:])
```

Масштабируем получившиеся выборки с обучающими признаками.

## Обучение и анализ моделей

Создадим функции для расчета RMSE и получения информации о метриках моделей.

```python
def ml_rmse(ml_target, ml_pred):
    return np.sqrt(mean_squared_error(ml_target, ml_pred))

ml_rmse_score = make_scorer(ml_rmse, greater_is_better=False)





def get_info_on_regression_results(_time, ml_target, ml_pred):
    
    print('RMSE: ', round(ml_rmse(ml_target, ml_pred), 2))
    print('MAE: ', round(mean_absolute_error(ml_target, ml_pred),2))
    print('R2_score: ', round(r2_score(ml_target, ml_pred),2))
    print('Время операции {} сек.'.format(round(_time, 0) ))   
    
    pass

def get_results(_time, best_score, best_params):
    
    print('Время подбора {} сек.'.format(round(_time, 0) ))
    print('Лучший результат: %s' % round(np.abs(gs_result.best_score_), 2))
    print('Лучшая модель с параметрами: %s' % gs_result.best_params_)
    
    pass
```
