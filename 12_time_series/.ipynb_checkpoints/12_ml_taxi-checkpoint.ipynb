# Прогноз заказов такси

**Описание проекта**

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. <br>
Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. <br>

Цель проекта - построить модель для предсказания количества заказов такси на следующий час.<br>
При это значение метрики RMSE на тестовой выборке должно быть не больше 48.

**Шаги выполнения проекта:**
- Загрузим данные и выполним их ресемплирование по одному часу.
- Проанализируем данные.
- Обучим разные модели с различными гиперпараметрами. 
- Сделаем тестовую выборку размером 10% от исходных данных.
- Проверим данные на тестовой выборке и сделаем выводы.

```python
import time
import math
import warnings
import numpy as np
import pandas as pd

import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller 

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.dummy import DummyRegressor


from sklearn.metrics import r2_score
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error


matplotlib.style.use('ggplot')

pd.set_option('display.max_rows', None)
pd.options.mode.chained_assignment = None
warnings.filterwarnings(action="ignore", message="Mean of empty slice")
warnings.filterwarnings("ignore")
```

```python
def get_info_on_dataset(_df):
    
    CONST_IN_PERCENTAGES = float(100.0)
    
    try:
        _df.info()
        df_review_info = pd.DataFrame({'Not_Valid_Values_in %': (_df.isna().sum()/len(_df))*CONST_IN_PERCENTAGES} )
        print("")
        display(df_review_info.sort_values(by='Not_Valid_Values_in %', ascending=False))
            
        print("")
        print("Number of duplicated rows in the dataset: {}".format(_df.duplicated().sum()))
            
    except:
        print("Can't provide review on given dataset")
    
    pass


def plot_info_on_numerical_column( _df, 
                                   _feature_title,
                                   _text_info = True,
                                   _bins = 50, 
                                   _figure_width = 20, 
                                   _figure_height = 12, 
                                   _fontsize=15,
                                   _number_of_decimals = 2):

    NCOLS_NUMBER_CONST = 2
    
    fig, ax = plt.subplots(ncols=NCOLS_NUMBER_CONST, figsize=(_figure_width, _figure_height))

    sns.histplot(data=_df, 
                 x=_feature_title, 
                 color="cornflowerblue", 
                 label=_feature_title, 
                 bins=_bins, 
                 kde=True,
                 ax=ax[0])
        
    mean_value = _df[_feature_title].mean()  
    median_value = _df[_feature_title].median() 
    
    ax[0].axvline(x=mean_value, color='black', lw = 2, ls = ':', 
                  label = "mean: {}".format(round(mean_value, _number_of_decimals)))
    ax[0].axvline(x=median_value, color='blue', lw = 2, ls = '-.', 
                  label = "median: {}".format(round(median_value, _number_of_decimals)))

    ax[0].legend(fontsize = '15', title_fontsize = "10", loc='upper right') 
        
    plt.tight_layout()
    
    Q1_QUANTILE_VALUE_CONST = float(0.25)
    Q3_QUANTILE_VALUE_CONST = float(0.75)
    IQR_MULT_DEFAULT_CONST = float(1.5)
    
    Q1 = _df[_feature_title].quantile(Q1_QUANTILE_VALUE_CONST)
    Q3 = _df[_feature_title].quantile(Q3_QUANTILE_VALUE_CONST)
    IQR = Q3 - Q1 

            
    outliers_low_bound = Q1 - IQR_MULT_DEFAULT_CONST*IQR
    outliers_upper_bound = Q3 + IQR_MULT_DEFAULT_CONST*IQR      
    
    sns.boxplot(_df[_feature_title], color="cornflowerblue", ax=ax[1])

    if _text_info:
        print("Q1 == {:.2f}, Q3 == {:.2f}, IQR == {:.2f}".format(Q1, Q3, IQR))
        print("[Q1 - 1.5*IQR] == {:.2f}, [Q3 + 1.5*IQR] == {:.2f}".format(outliers_low_bound, outliers_upper_bound))    

    plt.show()
    

def plot_df_coll_matrix(_df, 
                        _columns, 
                        _annotation = True,
                        _figure_height = 15, 
                        _figure_width = 15, 
                        _cell_annotation_font_size = 15,
                        _axiss_labels_font_size = 10):

    try:
        corr = _df[_columns].corr()
        fig, ax = plt.subplots(figsize=(_figure_height, _figure_width))           
        _pl = sns.heatmap(corr, 
                          ax=ax, 
                          annot=_annotation, 
                          linewidths=.5, 
                          fmt=".3f",
                          annot_kws={'size': _cell_annotation_font_size}, 
                          cmap='coolwarm',
                          square=True)
        
        _pl.set_yticklabels(ax.get_yticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
        _pl.set_xticklabels(ax.get_xticklabels(), size = _axiss_labels_font_size, fontweight = 'bold')
    except:
        print("Can't provide pairwise correlation plotting")    
    pass


def plot_time_series_native_data(_df, 
                                 _columns, 
                                _figure_width = 20, 
                                _figure_height = 12, 
                                _palette = 'Set2',
                                _axiss_labels_font_size = 12,
                                _fontsize=15,
                                _linewidth=2):
  
    fig, ax = plt.subplots(figsize=(_figure_width, _figure_height))
    
    _pl = sns.lineplot(data=_df[_columns], 
                       palette=_palette,
                       linewidth=_linewidth)
    
    _pl.set_yticklabels(_pl.get_yticks(), size = _axiss_labels_font_size)
   
    _pl.set_xlabel(_pl.get_xlabel(),fontsize=_axiss_labels_font_size, fontweight = 'bold')
    _pl.set_ylabel(_pl.get_ylabel(),fontsize=_axiss_labels_font_size, fontweight = 'bold')

    plt.setp(ax.get_legend().get_texts(), fontsize=_fontsize) 
    plt.setp(ax.get_legend().get_title(), fontsize=_fontsize)     
    
    pass


def plot_features_importances(_df,
                              _figure_width = 17, 
                              _figure_height = 10, 
                              _fontsize=15):
    
    fig, ax = plt.subplots(figsize=(_figure_width, _figure_height)) 

    _pl = sns.barplot(x="importances_value", 
                      y="features", 
                      data=_df, 
                      orient="h", 
                      order=_df.sort_values(by="importances_value", ascending=False).features.tolist(), 
                      color="c");

    _pl.set_yticklabels(_pl.get_yticklabels(), size = _fontsize)
    _pl.set_xticklabels(_pl.get_xticks(), size = _fontsize)

    _pl.set_xlabel(_pl.get_xlabel(),fontsize=_fontsize, fontweight = 'bold')
    _pl.set_ylabel(_pl.get_ylabel(),fontsize=_fontsize, fontweight = 'bold')
    
    pass
```

```python
_ROLLING_WINDOW_SIZE_CONST_VALUE_1 = 24*7

_CLOSE_TO_ZERO_VALUE_CONST = 10E-1
_FEATURES_LAG_VALUE_CONST = 24
_FEATURES_ROLLING_MEAN_LIST_VALUES_CONST = [6,12,24]

ALPHA_CONST_VALUE = float(.01)
ML_RND_GNR_SEED_VALUE_CONST = 424
ML_TEST_SUBSET_SIZE_VALUE_CONST = 0.1

_ML_TARGET_COLUMN_NAME_CONST = 'ml_target'

_CROSS_VALIDATION_FOLDS_NUMBER_CONST = 5
```

Похоже, что наш ряд является нестационарным, так как есть тренд и сезонность - зависимость от времени.

Проведем тест Дики-Фуллера, проверим стационарность процесса.

Сформулируем нулевую и альтернативные гипотезы:
- $H_0$ - процесс нестационарный
- $H_1$ - соответственно говорит об обратном
 
Используем критический уровень статистической значимости = 1% ($\alpha = 0.01$)


```python
ALPHA_CONST_VALUE = float(.01)

adfuller_test = adfuller(df_taxi_r_1h['num_orders'])
p_value = adfuller_test[1]
print('p-value = ', p_value)
if p_value < ALPHA_CONST_VALUE:
    print('p_value < alpha ==> нулевая гипотеза отвергается')
else:
    print('p_value > alpha ==> нулевая гипотеза не может быть отвергнута')
```

Вероятность (p-value) для данных о заказах такси выше 0,01. <br>
Нулевая гипотеза не может быть отвергнута. Наш процесс может быть нестационарным.<br>
Попробуем сделать логарифмирование, для сглаживания сезонности и проведем тест Дики-Фуллера еще раз.

```python
df_taxi_r_1h['num_orders_log'] = df_taxi_r_1h['num_orders'].map(
    lambda x: math.log(_CLOSE_TO_ZERO_VALUE_CONST) if x<=0 else math.log(x))
```

```python
adfuller_test = adfuller(df_taxi_r_1h['num_orders_log'])
p_value = adfuller_test[1]
print('p-value = ', p_value)
if p_value < ALPHA_CONST_VALUE:
    print('p_value < alpha ==> нулевая гипотеза отвергается')
else:
    print('p_value > alpha ==> нулевая гипотеза не может быть отвергнута')
```

Вероятность (p-value) для данных о заказах такси ниже 0,01. <br>
Мы отвергаем нулевую гипотезу. Наш процесс является стационарным.

```python
df_taxi_r_1h['mean_log'] = df_taxi_r_1h['num_orders_log'].rolling(_ROLLING_WINDOW_SIZE_CONST_VALUE_1).mean()
df_taxi_r_1h['median_log'] = df_taxi_r_1h['num_orders_log'].rolling(_ROLLING_WINDOW_SIZE_CONST_VALUE_1).median()
df_taxi_r_1h['std_log'] = df_taxi_r_1h['num_orders_log'].rolling(_ROLLING_WINDOW_SIZE_CONST_VALUE_1).std()
```

## Создание признаков и подготовка данных для моделей

Создаем признаки для обучения модели.

```python
df_taxi_features = df_taxi_r_1h[['num_orders_log']].copy()
```

```python
df_taxi_features['hour'] = df_taxi_features.index.hour
df_taxi_features['day'] = df_taxi_features.index.day
df_taxi_features['day_of_week'] = df_taxi_features.index.dayofweek

for _lag_value in range(1, _FEATURES_LAG_VALUE_CONST+1):
    df_taxi_features['f_lag_{}'.format(_lag_value)] = df_taxi_features['num_orders_log'].shift(_lag_value)
    
for _rolling_mean_value in _FEATURES_ROLLING_MEAN_LIST_VALUES_CONST:
    df_taxi_features['f_rolling_mean_{}'.format(_rolling_mean_value)] \
    = df_taxi_features['num_orders_log'].shift().rolling(_rolling_mean_value).mean()
```

```python
df_taxi_features = df_taxi_features.drop(['num_orders_log'], axis=1)
df_taxi_features[_ML_TARGET_COLUMN_NAME_CONST] = df_taxi_r_1h['num_orders']
df_taxi_features = df_taxi_features.dropna().reset_index(drop=True)
```

Удалили строки с пропущенными значениями в столбцах `lag`. Итого получилось 30 обучающих признаков.

На корреляционной матрице видим высокую корреляцию целевого признака и следующих столбцов: `f_lag_24`, `f_lag_1`, `f_rolling_mean_24`. <br>

## Подбор гиперпараметров, обучение и тестирование моделей

Создадим функции для расчета RMSE и получения информации о метриках моделей.

```python
def ml_rmse(ml_target, ml_pred):
    return np.sqrt(mean_squared_error(ml_target, ml_pred))

ml_rmse_score = make_scorer(ml_rmse, greater_is_better=False)


def get_info_on_regression_results(_time, ml_target, ml_pred):
    
    print('RMSE:     ', round(ml_rmse(ml_target, ml_pred), 2))
    print('MAE:      ', round(mean_absolute_error(ml_target, ml_pred),2))
    print('R2_score:  ', round(r2_score(ml_target, ml_pred),2))
    print('Время операции {} сек.'.format(round(_time, 0) ))       
    pass


def get_results(_time, best_score, best_params):
    
    print('Время подбора                 {} сек.'.format(round(_time, 0) ))
    print('Лучший результат:            ', round(np.abs(gs_result.best_score_), 2))
    print('Лучшая модель с параметрами: ', gs_result.best_params_)
    
    pass

    
results = pd.DataFrame()
results['model_name'] = ['LinearRegression','RandomForest','CatBoost','XGBoost','LGBM']
results['grid_search_time'] = pd.Series([])
results['rmse_best_score'] = pd.Series([])
results['train_time'] = pd.Series([])
results['rmse_train'] = pd.Series([])
results['test_time'] = pd.Series([])
results['rmse_test'] = pd.Series([])
```

Зададим кросс-валидацию для временного ряда

```python
time_series_cv = TimeSeriesSplit(n_splits=_CROSS_VALIDATION_FOLDS_NUMBER_CONST)
```

Приступим к обучению моделей.
